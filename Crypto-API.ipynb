```python
import pandas as pd
import requests
```

```python
def get_crypto_data(crypto_id):
    url = f"https://api.coingecko.com/api/v3/coins/{crypto_id}"
    response = requests.get(url)
    return response.json()
```

```python
crypto_data = get_crypto_data('bitcoin')
```

```python
crypto_data
```

```python
# Transforming Data
```

```python
def transform_crypto_data(crypto_data):
    current_price = crypto_data['market_data']['current_price']['usd']
    market_cap = crypto_data['market_data']['market_cap']['usd']
    volume = crypto_data['market_data']['total_volume']['usd']
    
    # Create a DataFrame
    data = {
        'Price (USD)': [current_price],
        'Market Cap (USD)': [market_cap],
        'Volume (USD)': [volume]
    }
    return pd.DataFrame(data)

transformed_data = transform_crypto_data(crypto_data)
print(transformed_data)
```
